## 1. Introduction to EDA

Beberapa pertanyaan muncul di pikiran ketika kita menemukan dataset baru. Daftar di bawah ini menjelaskan beberapa pertanyaan tersebut:-
- Bagaimana distribusi dataset?
- Apakah ada missing values, outliers, atau anomali dalam dataset?
- Apa asumsi yang mendasari dataset tersebut?
- Apakah ada hubungan antar variabel dalam dataset?
- Bagaimana cara memastikan bahwa dataset kita siap untuk dimasukkan ke dalam algoritme machine learning?
- Bagaimana cara memilih algoritma yang paling sesuai untuk dataset yang diberikan?

Jadi, bagaimana cara kita mendapatkan jawaban dari pertanyaan-pertanyaan di atas? 

Jawabannya adalah **Exploratory Data Analysis**. Analisis ini memungkinkan kita untuk menjawab semua pertanyaan di atas.

**Exploratory Data Analysis** atau EDA adalah langkah pertama yang sangat penting dalam menganalisis dataset baru. Tujuan utama dari EDA adalah untuk menganalisis data dalam distribusi, outliers, dan anomali dalam dataset. Hal ini memungkinkan kita untuk mengarahkan pengujian hipotesis secara spesifik, termasuk menganalisis data untuk menemukan distribusi data, karakteristik utamanya, mengidentifikasi pola dan visualisasi. Selain itu, EDA juga menyediakan alat untuk pembuatan hipotesis dengan memvisualisasikan dan memahami data melalui representasi grafis.


## 2. Distribution of a Variable

Variabel berarti jumlah objek yang dipertimbangkan sebagai sampel dalam percobaan. Ada tiga jenis distribusi variabel adalah sebagai berikut:

1. **Distribusi Univariat**
   Dalam distribusi univariat, hanya ada satu variabel yang dipertimbangkan. Distribusi ini adalah bentuk analisis yang paling sederhana karena hanya satu kuantitas yang berubah. Analisis ini tidak berurusan dengan penyebab atau hubungan. Tujuan utama dari analisis ini adalah untuk mendeskripsikan data dan menemukan pola yang ada di dalamnya. Kita dapat menggambarkan pola yang ditemukan dalam data univariat menggunakan tendensi sentral (rata-rata, median dan modus) dan dispersi (rentang, varians, standar deviasi, nilai maksimum dan minimum dan rentang interkuartil). Kita dapat memvisualisasikan data univariat menggunakan berbagai jenis bagan dan grafik, seperti tabel distribusi frekuensi, histogram, diagram batang, diagram lingkaran, dan poligon frekuensi.

2. **Distribusi Bivariat**  
   Jenis distribusi data ini melibatkan dua variabel yang berbeda. Analisis jenis data ini berkaitan dengan sebab dan hubungan dan analisis dilakukan untuk mengetahui hubungan di antara dua variabel. Contoh distribusi bivariat yang sangat umum adalah tinggi dan berat badan seseorang. Dengan demikian, analisis data bivariat melibatkan perbandingan, mengeksplorasi hubungan, menemukan penyebab dan penjelasan. Variabel-variabel ini sering kali diplot pada sumbu X dan Y pada grafik untuk pemahaman data yang lebih baik dan salah satu dari variabel-variabel ini bersifat independen sementara yang lain bersifat dependen.

   Jenis analisis bivariat yang umum termasuk menggambar scatter plot, analisis regresi, dan menemukan koefisien korelasi. Scatter plot digunakan untuk mengetahui apakah ada hubungan antara dua variabel. Analisis regresi adalah metode statistik untuk memperkirakan hubungan antar variabel. Analisis koefisien korelasi mengukur kekuatan dan arah hubungan linier antara dua variabel pada scatter plot.

3. **Distribusi Multivariat**
   Ketika kumpulan data melibatkan tiga variabel atau lebih, maka dikategorikan dalam distribusi multivariat. Analisis multivariat digunakan untuk mempelajari kumpulan data yang lebih kompleks. Biasanya tidak cocok untuk kumpulan data yang kecil. Ada berbagai macam teknik analisis untuk melakukan analisis multivariat. Pilihan teknik analisis tergantung pada dataset dan tujuan yang ingin dicapai. Beberapa contoh teknik analisis multivariat adalah additive tree, cluster analysis, correspondence analysis, factor analysis, MANOVA (multivariate analysis of variance), multidimensional scaling, multiple regression analysis, principal component analysis and redundancy analysis.

## 10. Univariate analysis

### Ukuran Tendensi Sentral dan Dispersi

**Tendensi Sentral** berarti nilai pusat yang menggambarkan distribusi probabilitas. Hal ini juga dapat disebut sebagai pusat atau lokasi distribusi. Ukuran tendensi sentral yang paling umum adalah rata-rata, median, dan modus. Untuk distribusi yang miring atau ketika ada kekhawatiran tentang outliers, median mungkin lebih disukai. Jadi, median adalah ukuran yang lebih kuat daripada mean.

**Dispersi** adalah indikator seberapa jauh distribusi data dari pusat. Ukuran dispersi yang paling umum adalah varians, deviasi standar, dan rentang interkuartil (IQR). Varians adalah ukuran standar penyebaran. Deviasi standar adalah akar kuadrat dari varians. 
Interquartile Range (IQR) dihitung dengan menggunakan batas-batas data yang terletak di antara kuartil ke-1 dan ke-3. Jadi, IQR dapat dihitung sebagai IQR = Q3 - Q1. IQR adalah ukuran penyebaran yang kuat.

````python
df['Absenteeism time in hours'].describe()
````

**Interpretation**

- Nilai count, min dan max mewakili jumlah hitungan, nilai minimum dan maksimum dari variabel target.  
- Ukuran tendensi sentral diberikan oleh mean (6,924324) dan median (nilai 50% - 3,00).  
- Ukuran dispersi diberikan oleh deviasi standar yang diberikan oleh std (13.330998).  
- Nilai 25%, 50% dan 75% menunjukkan persentil yang sesuai. Persentil ke-50 menunjukkan median dari distribusi.  
- IQR adalah perbedaan antara persentil ke-75 dan ke-25. Oleh karena itu, IQR = 8,00 - 2,00 = 6,00  

### Measures of Shape

Kita telah melihat ukuran tendensi sentral data (rata-rata dan median) dan penyebaran data (standar deviasi (std), IQR, nilai minimum (min) dan maksimum (maks)). Kuantitas ini hanya dapat digunakan untuk variabel kuantitatif bukan untuk variabel kategorikal.

Sekarang, kita akan melihat ukuran-ukuran bentuk distribusi. Ada dua ukuran statistik yang dapat memberi tahu kita tentang bentuk distribusi. Ukuran-ukuran ini adalah **skewness** dan **kurtosis**. Ukuran-ukuran ini dapat digunakan untuk menyampaikan informasi tentang bentuk distribusi kumpulan data.

### Skewness

Skewness adalah ukuran simetri distribusi suatu data dalam menunjukkan bentuk distribusi data. Proses ini digunakan untuk mengartikan tidak adanya simetri dari rata-rata kumpulan data atau karakteristik penyimpangan dari rata-rata. Ini digunakan 


### Negative skewness

Nilai negatif untuk skewness menunjukkan kemiringan negatif. Dalam hal ini, data condong atau ekor ke kiri, dimana ekor kiri lebih panjang relatif terhadap ekor kanan. Nilai data dapat meluas lebih jauh ke kiri tetapi terkonsentrasi di sebelah kanan. Jadi, ada ekor yang panjang dan distorsi disebabkan oleh nilai yang sangat kecil yang menarik rata-rata ke bawah sehingga kurang dari median.

**Mean < Median < Mode**



### Zero skewness

Zero skewness berarti nilai kemiringan nol, berarti kumpulan data tersebut simetris. Kumpulan data dikatakan simetris jika membentuk lonceng, dimana bagian kiri dan kanan memiliki bentuk yang sama dengan simetris sempurna. 

**Mean = Median = Mode**



### Positive skewness

Nilai positif untuk skewness menunjukkan kemiringan positif. Kumpulan data condong atau ekor ke kanan, dimana ekor kanan lebih panjang daripada ekor kiri. Nilai data terkonsentrasi di sebelah kanan. Jadi, ada ekor yang panjang ke 
ke kanan yang disebabkan oleh nilai yang sangat besar dalam menarik rata-rata ke atas sehingga lebih besar dari median.

**Mean > Median > Mode**



### Reference range on skewness values

Aturan praktis untuk nilai skewness adalah:  
- Jika skewness berada di antara -0,5 dan 0,5, data cukup simetris.  
- Jika kemiringan antara -1 dan -0,5 atau antara 0,5 dan 1, data cukup miring.  
- Jika skewness kurang dari -1 atau lebih besar dari 1, data tersebut sangat miring.  


````python
df['Absenteeism time in hours'].skew()
````

**Interpretation of skewness**

Skewness dari variabel target ternyata lebih besar dari +1. Jadi, kita dapat menyimpulkan bahwa variabel target sangat condong ke arah positif. 

Kita dapat mengkonfirmasi hal ini dengan membuat plot diagram sebagai berikut:

**Conclusion**  

Plot di atas menegaskan bahwa variabel target `Absenteeism time in hours` sangat condong ke arah positif.

### Findings of univariate analysis

Koefisien korelasi berkisar antara -1 hingga +1.  

- Ketika mendekati +1, ini menandakan bahwa ada korelasi positif yang kuat. Jadi, kita dapat melihat bahwa ada korelasi positif yang kecil antara `Absenteeism time in hours` dan `Height`.  

- Jika mendekati -1, ini berarti ada korelasi negatif yang kuat. Jadi, ada korelasi negatif yang kecil antara `Absenteeism time in hours` dan `Reason for absence`.  

- Jika mendekati 0, berarti tidak ada korelasi. Jadi, tidak ada korelasi antara `Absenteeism time in hours` dan `Seasons`.
